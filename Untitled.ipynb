```python
import numpy as np
from mofa_subtypes.entry_point import entry_point_subtypes

rng = np.random.default_rng(0)
N = 30
X1 = rng.normal(size=(N, 15))
X2 = rng.normal(size=(N, 10))

data = [
    [X1],  # view 1, group 1
    [X2],  # view 2, group 1
]

ep = entry_point_subtypes()

# likelihoods goes HERE in mofapy2
ep.set_data_matrix(
    data,
    likelihoods=["gaussian", "gaussian"],
    views_names=["v1", "v2"],
    groups_names=["g1"],
)

# model options: no likelihoods kwarg
ep.set_model_options(factors=3)

ep.set_train_options(iter=20, verbose=True)

ep.set_subtypes_options(C=4)

ep.build()
ep.run()
ep.save("smoke_model.hdf5")
```

```python
import inspect
print(inspect.signature(ep.set_data_matrix))
print(inspect.signature(ep.set_model_options))
```

```python
ep.build()

nodes = list(ep.model.getNodes().keys())
base  = ep.train_opts.get("schedule", [])

# keep base order where possible, then append any missing nodes
schedule = [n for n in base if n in nodes] + [n for n in nodes if n not in base]
ep.train_opts["schedule"] = schedule

ep.run()
```

```python
ep.set_subtypes_options(C=4)
ep.set_train_options(iter=20, verbose=True)
ep.build()

ep.train_opts["schedule"] = list(ep.model.getNodes().keys())
ep.run()
```

```python
ep.build()

nodes = list(ep.model.getNodes().keys())
sched = ep.train_opts.get("schedule", [])

nodes_set = set(nodes)
sched_set = set(sched)

print("N nodes:", len(nodes))
print("N sched:", len(sched))

print("Missing in schedule (in model, not in schedule):", sorted(nodes_set - sched_set))
print("Extra in schedule (in schedule, not in model):", sorted(sched_set - nodes_set))
```
